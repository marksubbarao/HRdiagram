```python
from astropy.io import ascii
from astropy.table import Table
from astropy import units as u
```

```python
starTable = ascii.read('hygdata_v3.csv')
```

```python
starTable.show_in_browser(jsviewer=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn
import numpy as np
from astropy.coordinates import Angle
fig = plt.figure (figsize=(13,3))
ax = fig.add_subplot(121,projection="mollweide")
ax.grid(True)
ax.set_title("GAMA Survey Sky Coverage")
ax.scatter(Angle(starTable['ra'],u.hr).wrap_at(180.*u.deg).radian,Angle(starTable['dec'],u.deg).radian,s=1)
a2 = fig.add_subplot(122)
a2.hist(starTable['dist'],bins=100)
a2.set_title("redshift distribution")
a2.set_xlabel("z")
a2.set_ylabel("N")
```

```python
fig = plt.figure (figsize=(13,6))
ax = fig.add_subplot(111,projection="mollweide")
plt.scatter(Angle(starTable['ra'],u.hr).wrap_at(180.*u.deg).radian,Angle(starTable['dec'],u.deg).radian,s=(5.0-starTable['absmag']))
```
